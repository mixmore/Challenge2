```python

def insert_contract():
   
    with open('D:/Blockchain/layer1.csv', 'r') as csv:
        last_index = [[x.strip() for x in line.strip().split(',')] for line in csv.readlines()][-1][0]   
        csv.close()
    

    
    with open('D:/Blockchain/layer1.csv', 'r') as csv1:
        last_hash1 = [[x.strip() for x in line.strip().split(',')] for line in csv1.readlines()][-1][1]   
        csv1.close() 
   
    index1 = int(last_index) + 1
    hash1 = last_hash1
    
    data1 = cText1.get(1.0, "end-1c") 
    data2 = cText2.get(1.0, "end-1c") 

    
    hash_string = str(index1) + str(date.datetime.now()) + str(data1) + str(data2) + str(hash1)
    new_hash= hashlib.sha256(hash_string.encode()).hexdigest()

    list_data=[str(index1),str(new_hash),str(date.datetime.now()),str(data1),str(data2), str(hash1)]

    
    
    
    with open('D:/Blockchain/layer1.csv', 'a', newline='') as f_object:  
        writer_object = writer(f_object)
        writer_object.writerow(list_data)  
        f_object.close()
      

    
def contract_search():
    output1 = cText3.get(1.0, "end-1c") 
    
    with open('D:/Blockchain/layer1.csv') as file_obj:
      
         reader_obj = csv.reader(file_obj)
         for row in reader_obj:
            if row[1] == output1:
                #value = int(row[1])
                value = str(row[4])
                
    #cText4.insert_contract(INSERT, value)
    cText4.delete("1.0","end")
    cText4.insert(END, value)
    #cText4.insert(0, "lol")
    #print(value)

    
def contract_view():
    cpt.show()
    cpt.redraw()
    cpt.importCSV('D:/Blockchain/layer1.csv')
        


def excute():
    code = cText4.get(1.0, "end-1c") 
    #result = eval(code)
    #Print(code)
    exec(code)
    #result = eval(code)
    #result_text.insert(INSERT, str(result))
    #print(code)                    
        
```
